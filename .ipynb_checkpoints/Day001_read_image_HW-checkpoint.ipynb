思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# matplotlib RGB，opencv BGR

img_path = './data/lena.png'

# 以彩色圖片的方式載入
img = cv2.imread(img_path, cv2.IMREAD_COLOR) # img RGB
zeros = np.zeros(img.shape[:2], dtype="uint8")
B, G, R = cv2.split(img)
```

```python
def cv2plt(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
f = plt.figure(figsize=(12, 6))

f.add_subplot(1, 4, 1)
plt.imshow(cv2plt(cv2.merge([B, zeros, zeros]))) #we needs to change channel, because plt display by RGB 

f.add_subplot(1, 4, 2)
plt.imshow(cv2plt(cv2.merge([zeros, G, zeros])))

f.add_subplot(1, 4, 3)
plt.imshow(cv2plt(cv2.merge([zeros, zeros, R])))

f.add_subplot(1, 4, 4)
plt.imshow(cv2plt(cv2.merge([B, G, R])))

plt.show(block=True)
```
